#### Evaluate different models

```python
from traintest import ClassificationModel2, add_to_json_array

metric_results = []
```

## Stemming + CountVectorizer

```python
import joblib

enron_stemmed_countvec = joblib.load("./preprocess/enron_stemmed_countvec.pkl")
```

```python
enron_stemmed_countvec_X, enron_y = enron_stemmed_countvec["features"], enron_stemmed_countvec["labels"]
print(enron_stemmed_countvec_X.shape)

del enron_stemmed_countvec
```

```python
clsml2 = ClassificationModel2(enron_stemmed_countvec_X, enron_y, "enron_stemmed_countvec")
clsml2.split()

m_r = clsml2.train_and_evaluate_models(
    valid_size=0.1,
    max_workers=4,
    nomlp=True,
    threading=True,
    save_plot=True
)

metric_results.extend(m_r)

del enron_stemmed_countvec_X, enron_y, m_r
```

```python
del clsml2
```

```python
metric_results
```
